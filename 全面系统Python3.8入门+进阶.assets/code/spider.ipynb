```python
from urllib import request
import re


class Spider():
    result = []

    def __init__(self, url) -> None:
        self.url = url

    def __fetch_content(self):
        r = request.urlopen(self.url)
        return str(r.read(), encoding='utf8')

    def __analysis(self, html_str):
        reg_main = '<main class="container">[\s\S]*?</main>'
        container = re.findall(reg_main, html_str)
        # print(container)

        reg_div = '<div class="c-single-text-ellipsis">[\s\S]*?</div>'
        reg_a = 'class="c-single-text-ellipsis[\s\S]*?</a>'
        reg_a_sub = 'class="c-single-text-ellipsis[\s\S]*?>'

        def myPrint(container, reg_div):
            for text_div in container:
                div_container = re.findall(reg_div, text_div)
                # print(div_container)
                for div in div_container:
                    r_div = re.sub(r'<[^>]+>', "", div)
                    r_div = re.sub(reg_a_sub, "", r_div)
                    # print(r_div.strip())
                    self.result.append(r_div.strip())
        myPrint(container, reg_a)

    def __sortResult(self, str_list):
        # print(str_list)
        return sorted(str_list, key=self.__sort_seed, reverse=True)

    def __sort_seed(self, text):
        # print(text[0])
        return text[0]

    def go(self):
        html_str = self.__fetch_content()
        self.__analysis(html_str)
        return self.__sortResult(self.result)


spider = Spider('https://top.baidu.com/board?platform=pc&sa=pcindex_entry')
result = spider.go()
for index in range(0, len(result)):
    print(index + 1, result[index])
```
